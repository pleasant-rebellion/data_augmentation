```python
import pandas as pd

vanila_dataset = pd.read_csv("train_dataset_modified_1-3special_char_11.csv", encoding='utf-8')
```

```python
vanila_dataset
```

```python
chatbot_dataset = pd.DataFrame()
chatbot_dataset['Q'] = vanila_dataset['subject_entity'] + ' / ' + vanila_dataset['object_entity']
chatbot_dataset['A'] = vanila_dataset['label']
chatbot_dataset['label'] = int()

```

```python
import pickle
with open('/opt/ml/dict_label_to_num.pkl', 'rb') as f:
	data = pickle.load(f)
```

```python
for i in range(len(vanila_dataset)) :
    label = vanila_dataset['label'][i]
    chatbot_dataset['label'][i] = data[f'{label}']
```

```python
chatbot_dataset
```

```python
chatbot_dataset.to_csv("klue_chatbot.csv", encoding = 'utf-8')
```
