```python
import pandas as pd
dataset_t = pd.read_csv('train_dataset_modified_1-2hanja.csv', index_col=0, encoding='utf-8')
dataset_t
```

```python
data= dataset_t
        
sub_entity, sub_type= [], []
obj_entity, obj_type= [], []
sub_idx, obj_idx= [], []
sentence= []

"""preprocess"""
for i, [x, y, z] in enumerate(zip(data['subject_entity'], data['object_entity'], data['sentence'])):
    sub_typ= x[1:-1].split(':')[-1].split('\'')[-2]
    obj_typ= y[1:-1].split(':')[-1].split('\'')[-2]
            
    for idx_i in range(len(x)):
        if x[idx_i: idx_i+ 9]== 'start_idx':
            sub_start= int(x[idx_i+12:].split(',')[0].strip())
        if x[idx_i: idx_i+7]== 'end_idx':
            sub_end= int(x[idx_i+10:].split(',')[0].strip())
                
        if y[idx_i: idx_i+ 9]== 'start_idx':
            obj_start= int(y[idx_i+12:].split(',')[0].strip())
        if y[idx_i: idx_i+7]== 'end_idx':
            obj_end= int(y[idx_i+10:].split(',')[0].strip())
            
    sub_i= [sub_start, sub_end]
    obj_i= [obj_start, obj_end]

    sub_entity.append(z[sub_i[0]: sub_i[1]+1])
    obj_entity.append(z[obj_i[0]: obj_i[1]+1])
    sub_type.append(sub_typ); sub_idx.append(sub_i)
    obj_type.append(obj_typ); obj_idx.append(obj_i)
    
    # if sub_i[0] < obj_i[0]:
    #     z= z[:sub_i[0]] + '@*'+sub_typ+'*'+ z[sub_i[0]: sub_i[1]+1] + '@' + z[sub_i[1]+1:]
    #     z= z[:obj_i[0]+7] + '#^'+ obj_typ +'^'+ z[obj_i[0]+7: obj_i[1]+8]+ '#'+ z[obj_i[1]+8:]
    # else:
    #     z= z[:obj_i[0]] + '#^'+ obj_typ +'^'+ z[obj_i[0]: obj_i[1]+1]+ '#' + z[obj_i[1]+1:]
    #     z= z[:sub_i[0]+7] + '@*'+sub_typ+'*' + z[sub_i[0]+7: sub_i[1]+8] + '@' + z[sub_i[1]+8:]
        
    sentence.append(z)

test_df= pd.DataFrame({'id': data['id'], 'sentence' : sentence, 'subject_entity': sub_entity, 'object_entity': obj_entity,
                        'subject_type': sub_type, 'object_type': obj_type, 'label': data['label'],
                        'subject_idx': sub_idx, 'object_idx': obj_idx})
```

```python
# 온점, 반점 살리기
# 따옴표는 모두 통일시키기
# 그외는 모두 삭제
# 띄어쓰기는 모두 1개로 만들어버리기

import re
from tqdm.notebook import tqdm

unk_list = ['㈜','(주)','(사)','㈔','(재)','↔','–', '€', 'Ⓐ', '☎', '。', '˘',  '́', '⸱', '∞', '･', '⋅', '⟪', '⟫',
            '～','£','°', '˹', '˼', '▴','·', '☏', '⁺', '®', '、', '？', '／','《','》','㎞','㎏','↔','·','★','☆',
            '़', 'ी', 'ु', '्', 'ा', '㈔', '▵', '«', '»', '′', '₫', '㎿', '⌜', '⌟', '±','<','>','〈','〉',':',';',
           '▵','▵','~','～','∼','˹','˼','«','»','⌜','⌟',':','㎿','／','％','%','-','=','+','#','/','?','？',':','：','^','$',
            '@','*','※','~','&','%','ㆍ','!','』','|','[',']','<','>','…','》']

따옴표리스트 = ["'",'＇','‘','`','’','`','“','"','”', '’', '‘']

괄호열기리스트 = ['(','（','(','(',]
괄호닫기리스트 = [')', '）', ')',')',]

for i in tqdm(range(len(dataset_1_4))) :
    for j in unk_list :
        if j in dataset_1_4['sentence'][i] :
            dataset_1_4['sentence'][i] = dataset_1_4['sentence'][i].replace(j, " ",6)
        if j in dataset_1_4['subject_entity'][i] : 
            dataset_1_4['subject_entity'][i] = dataset_1_4['subject_entity'][i].replace(j, " ",6)
        if j in dataset_1_4['object_entity'][i] : 
            dataset_1_4['object_entity'][i] = dataset_1_4['object_entity'][i].replace(j, " ",6)
    
    for k in 따옴표리스트 :
        if k in dataset_1_4['sentence'][i] : 
            dataset_1_4['sentence'][i] = dataset_1_4['sentence'][i].replace(k, '"',3)
        if j in dataset_1_4['subject_entity'][i] : 
            dataset_1_4['subject_entity'][i] = dataset_1_4['subject_entity'][i].replace(k, '"',2)
        if j in dataset_1_4['object_entity'][i] : 
            dataset_1_4['object_entity'][i] = dataset_1_4['object_entity'][i].replace(k, '"',2)
    
    for x in 괄호열기리스트 :
        if j in dataset_1_4['sentence'][i] :
            dataset_1_4['sentence'][i] = dataset_1_4['sentence'][i].replace(x, "(",4)
        if j in dataset_1_4['subject_entity'][i] : 
            dataset_1_4['subject_entity'][i] = dataset_1_4['subject_entity'][i].replace(x, "(",2)
        if j in dataset_1_4['object_entity'][i] : 
            dataset_1_4['object_entity'][i] = dataset_1_4['object_entity'][i].replace(x, "(",2)
    
    for y in 괄호닫기리스트 : 
        if j in dataset_1_4['sentence'][i] :
            dataset_1_4['sentence'][i] = dataset_1_4['sentence'][i].replace(y, ")",4)
        if j in dataset_1_4['subject_entity'][i] : 
            dataset_1_4['subject_entity'][i] = dataset_1_4['subject_entity'][i].replace(y, ")",2)
        if j in dataset_1_4['object_entity'][i] : 
            dataset_1_4['object_entity'][i] = dataset_1_4['object_entity'][i].replace(y, ")",2)
    
    dataset_1_4['sentence'][i] = re.sub('[-=+#/\?:^$@*\※~&%ㆍ!』\\|\[\]\<\>\…》·]',' ', dataset_1_4['sentence'][i])
    dataset_1_4['subject_entity'][i] = re.sub('[-=+#/\?:^$@*\※~&%ㆍ!』\\|\[\]\<\>\…》·]',' ', dataset_1_4['subject_entity'][i])
    dataset_1_4['object_entity'][i] = re.sub('[-=+#/\?:^$@*\※~&%ㆍ!』\\|\[\]\<\>\…》·]',' ', dataset_1_4['object_entity'][i])
        
    if "'" in dataset_1_4['sentence'][i] :
        dataset_1_4['sentence'][i] = dataset_1_4['sentence'][i].replace("'", '"',4)
    if "'" in dataset_1_4['subject_entity'][i] : 
        dataset_1_4['subject_entity'][i] = dataset_1_4['subject_entity'][i].replace("'", '"',4)
    if "'" in dataset_1_4['object_entity'][i] : 
        dataset_1_4['object_entity'][i] = dataset_1_4['object_entity'][i].replace("'", '"',4)
        
    if '""' in dataset_1_4['sentence'][i] :
        dataset_1_4['sentence'][i] = dataset_1_4['sentence'][i].replace('""', '',4)
    if '""' in dataset_1_4['subject_entity'][i] :
        dataset_1_4['subject_entity'][i] = dataset_1_4['subject_entity'][i].replace('""', '',4)
    if '""' in dataset_1_4['object_entity'][i] : 
        dataset_1_4['object_entity'][i] = dataset_1_4['object_entity'][i].replace('""', '',4)
        
    # 두칸 띄어쓰기 이상 존재할 때
    dataset_1_4['sentence'][i] = re.sub(' +', ' ', dataset_1_4['sentence'][i])
    dataset_1_4['subject_entity'][i] = re.sub(' +', ' ', dataset_1_4['subject_entity'][i])
    dataset_1_4['object_entity'][i] = re.sub(' +', ' ', dataset_1_4['object_entity'][i])

    dataset_1_4['sentence'][i] = dataset_1_4['sentence'][i].strip()
    dataset_1_4['subject_entity'][i] = dataset_1_4['subject_entity'][i].strip()
    dataset_1_4['object_entity'][i] = dataset_1_4['object_entity'][i].strip()
```

## 1-4. 괄호 안 내용 지우기 전처리
- 괄호 안의 내용이 유의미하다면 지우지 않고, 그렇지 않다면 지운다.
- 괄호 불균형인 것들은 그냥 쳐낸다.
- 한자 전처리는 하지 않음!
- 특수문자는 처리하지 말자!

```python
import pandas as pd
pd.set_option('mode.chained_assignment',  None)
vanila_dataset_1_4 = pd.read_csv("train_dataset_modified1-1.csv", index_col=0, encoding='utf-8', )
vanila_dataset_1_4
```

```python
data= vanila_dataset_1_4
        
sub_entity, sub_type= [], []
obj_entity, obj_type= [], []
sub_idx, obj_idx= [], []
sentence= []

"""preprocess"""
for i, [x, y, z] in enumerate(zip(data['subject_entity'], data['object_entity'], data['sentence'])):
    sub_typ= x[1:-1].split(':')[-1].split('\'')[-2]
    obj_typ= y[1:-1].split(':')[-1].split('\'')[-2]
            
    for idx_i in range(len(x)):
        if x[idx_i: idx_i+ 9]== 'start_idx':
            sub_start= int(x[idx_i+12:].split(',')[0].strip())
        if x[idx_i: idx_i+7]== 'end_idx':
            sub_end= int(x[idx_i+10:].split(',')[0].strip())
                
        if y[idx_i: idx_i+ 9]== 'start_idx':
            obj_start= int(y[idx_i+12:].split(',')[0].strip())
        if y[idx_i: idx_i+7]== 'end_idx':
            obj_end= int(y[idx_i+10:].split(',')[0].strip())
            
    sub_i= [sub_start, sub_end]
    obj_i= [obj_start, obj_end]

    sub_entity.append(z[sub_i[0]: sub_i[1]+1])
    obj_entity.append(z[obj_i[0]: obj_i[1]+1])
    sub_type.append(sub_typ); sub_idx.append(sub_i)
    obj_type.append(obj_typ); obj_idx.append(obj_i)
        
    sentence.append(z)

dataset_1_4= pd.DataFrame({'id': data['id'], 'sentence' : sentence, 'subject_entity': sub_entity, 'object_entity': obj_entity,
                        'subject_type': sub_type, 'object_type': obj_type, 'label': data['label'],
                        'subject_idx': sub_idx, 'object_idx': obj_idx})
dataset_1_4
```

```python
from tqdm.notebook import tqdm
broken_cnt = []
for x in tqdm(range(len(dataset_1_4))) :
    modified_sentence = ""
    temp = ""
    괄호cnt = 0
        
    # 수정해야함 해당 idx 에 해당하면 넘어가는걸로다가 하자
    subject_len = len(dataset_1_4['subject_entity'][x])
    object_len = len(dataset_1_4['object_entity'][x])
    
    subject_start = dataset_1_4['sentence'][x].find(dataset_1_4['subject_entity'][x])
    object_start = dataset_1_4['sentence'][x].find(dataset_1_4['object_entity'][x])
    
    dataset_1_4['subject_idx'][x][0] =subject_start
    dataset_1_4['subject_idx'][x][1] = subject_start + subject_len -1
    dataset_1_4['object_idx'][x][0] = object_start
    dataset_1_4['object_idx'][x][1] = object_start + object_len -1
    
    if dataset_1_4['sentence'][x].count('(') !=  dataset_1_4['sentence'][x].count(')') :
        broken_cnt.append(x)
        pass

    else : 
        for i in range(len(dataset_1_4['sentence'][x])) :
            # 만약 괄호 시작되면?
            if dataset_1_4['sentence'][x][i] == "(" :
                괄호cnt += 1
                temp += dataset_1_4['sentence'][x][i]

            # 만약 괄호가 끝나면?
            elif dataset_1_4['sentence'][x][i] == ")" :
                괄호cnt -= 1
                temp += dataset_1_4['sentence'][x][i]
                # 남은 것 없이 괄호가 완전히 끝난경우?
                if 괄호cnt == 0 :
                    temp_len = len(temp)
                    start_temp_num = i - len(temp)
                    end_temp_num = i
                    
                    if (dataset_1_4['subject_entity'][x] in temp) or (dataset_1_4['object_entity'][x] in temp) :
                        modified_sentence += temp
                        temp = ""
                    elif (dataset_1_4['subject_idx'][x][0] <= start_temp_num) & (start_temp_num <= dataset_1_4['subject_idx'][x][1]) :
                        modified_sentence += temp
                        temp = ""
                    elif (dataset_1_4['object_idx'][x][0] <= start_temp_num) & (start_temp_num <= dataset_1_4['object_idx'][x][1]) :
                        modified_sentence += temp
                        temp = ""
                    elif (start_temp_num <= dataset_1_4['subject_idx'][x][0] ) & (dataset_1_4['subject_idx'][x][0] <= end_temp_num) :
                        modified_sentence += temp
                        temp = "" 
                    elif (start_temp_num <= dataset_1_4['subject_idx'][x][1] ) & (dataset_1_4['subject_idx'][x][1] <= end_temp_num) :
                        modified_sentence += temp
                        temp = "" 
                    elif (start_temp_num <= dataset_1_4['object_idx'][x][0] ) & (dataset_1_4['object_idx'][x][0] <= end_temp_num) :
                        modified_sentence += temp
                        temp = ""  
                    elif (start_temp_num <= dataset_1_4['object_idx'][x][1] ) & (dataset_1_4['object_idx'][x][1] <= end_temp_num) :
                        modified_sentence += temp
                        temp = "" 
                    else : 
                        temp = ""
                elif 괄호cnt < 0 :
                    broken_cnt.append(x)
                    
            # 괄호에 해당되지 않는 문자열이면
            else :
                if 괄호cnt != 0 :
                    temp += dataset_1_4['sentence'][x][i]
                else :
                    modified_sentence += dataset_1_4['sentence'][x][i] 
    
    dataset_1_4['sentence'][x] = modified_sentence

broken_cnt = set(broken_cnt)
# broken_cnt 해당되는 문장들 다 지우기
for i in broken_cnt : 
    dataset_1_4 = dataset_1_4.drop(i)
dataset_1_4['id'] = [i for i in range(len(dataset_1_4))]
# dataset_1_4 = dataset_1_4[['id','sentence', 'subject_entity','object_entity','object_type', 'subject_type', 'label','subject_idx','object_idx']]
dataset_1_4.reset_index(drop=False, inplace=True)
dataset_1_4.drop(['index'], axis=1, inplace=True)

dataset_1_4  
```

```python
data= test_dataset_modified
        
sub_entity, sub_type= [], []
obj_entity, obj_type= [], []
sub_idx, obj_idx= [], []
sentence= []

"""preprocess"""
for i, [x, y, z] in enumerate(zip(data['subject_entity'], data['object_entity'], data['sentence'])):
    sub_typ= x[1:-1].split(':')[-1].split('\'')[-2]
    obj_typ= y[1:-1].split(':')[-1].split('\'')[-2]
            
    for idx_i in range(len(x)):
        if x[idx_i: idx_i+ 9]== 'start_idx':
            sub_start= int(x[idx_i+12:].split(',')[0].strip())
        if x[idx_i: idx_i+7]== 'end_idx':
            sub_end= int(x[idx_i+10:].split(',')[0].strip())
                
        if y[idx_i: idx_i+ 9]== 'start_idx':
            obj_start= int(y[idx_i+12:].split(',')[0].strip())
        if y[idx_i: idx_i+7]== 'end_idx':
            obj_end= int(y[idx_i+10:].split(',')[0].strip())
            
    sub_i= [sub_start, sub_end]
    obj_i= [obj_start, obj_end]

    sub_entity.append(z[sub_i[0]: sub_i[1]+1])
    obj_entity.append(z[obj_i[0]: obj_i[1]+1])
    sub_type.append(sub_typ); sub_idx.append(sub_i)
    obj_type.append(obj_typ); obj_idx.append(obj_i)
    
    # if sub_i[0] < obj_i[0]:
    #     z= z[:sub_i[0]] + '@*'+sub_typ+'*'+ z[sub_i[0]: sub_i[1]+1] + '@' + z[sub_i[1]+1:]
    #     z= z[:obj_i[0]+7] + '#^'+ obj_typ +'^'+ z[obj_i[0]+7: obj_i[1]+8]+ '#'+ z[obj_i[1]+8:]
    # else:
    #     z= z[:obj_i[0]] + '#^'+ obj_typ +'^'+ z[obj_i[0]: obj_i[1]+1]+ '#' + z[obj_i[1]+1:]
    #     z= z[:sub_i[0]+7] + '@*'+sub_typ+'*' + z[sub_i[0]+7: sub_i[1]+8] + '@' + z[sub_i[1]+8:]
        
    sentence.append(z)

test_dataset_modified= pd.DataFrame({'id': data['id'], 'sentence' : sentence, 'subject_entity': sub_entity, 'object_entity': obj_entity,
                        'subject_type': sub_type, 'object_type': obj_type, 'label': data['label'],
                        'subject_idx': sub_idx, 'object_idx': obj_idx})
test_dataset_modified
```

```python
from tqdm.notebook import tqdm
broken_cnt = []
for x in tqdm(range(len(test_dataset_modified))) :
    modified_sentence = ""
    temp = ""
    괄호cnt = 0
        
    # 수정해야함 해당 idx 에 해당하면 넘어가는걸로다가 하자
    subject_len = len(test_dataset_modified['subject_entity'][x])
    object_len = len(test_dataset_modified['object_entity'][x])
    
    subject_start = test_dataset_modified['sentence'][x].find(test_dataset_modified['subject_entity'][x])
    object_start = test_dataset_modified['sentence'][x].find(test_dataset_modified['object_entity'][x])
    
    test_dataset_modified['subject_idx'][x][0] =subject_start
    test_dataset_modified['subject_idx'][x][1] = subject_start + subject_len -1
    test_dataset_modified['object_idx'][x][0] = object_start
    test_dataset_modified['object_idx'][x][1] = object_start + object_len -1
    
    if test_dataset_modified['sentence'][x].count('(') !=  test_dataset_modified['sentence'][x].count(')') :
        broken_cnt.append(x)
        pass

    else : 
        for i in range(len(test_dataset_modified['sentence'][x])) :
            # 만약 괄호 시작되면?
            if test_dataset_modified['sentence'][x][i] == "(" :
                괄호cnt += 1
                temp += test_dataset_modified['sentence'][x][i]

            # 만약 괄호가 끝나면?
            elif test_dataset_modified['sentence'][x][i] == ")" :
                괄호cnt -= 1
                temp += test_dataset_modified['sentence'][x][i]
                # 괄호가 완전히 끝난경우?
                if 괄호cnt == 0 :
                    temp_len = len(temp)
                    start_temp_num = i - len(temp)
                    end_temp_num = i
                    
                    if (test_dataset_modified['subject_entity'][x] in temp) or (test_dataset_modified['object_entity'][x] in temp) :
                        modified_sentence += temp
                        temp = ""
                    elif (test_dataset_modified['subject_idx'][x][0] <= start_temp_num) & (start_temp_num <= test_dataset_modified['subject_idx'][x][1]) :
                        modified_sentence += temp
                        temp = ""
                    elif (test_dataset_modified['object_idx'][x][0] <= start_temp_num) & (start_temp_num <= test_dataset_modified['object_idx'][x][1]) :
                        modified_sentence += temp
                        temp = ""
                    elif (start_temp_num <= test_dataset_modified['subject_idx'][x][0] ) & (test_dataset_modified['subject_idx'][x][0] <= end_temp_num) :
                        modified_sentence += temp
                        temp = "" 
                    elif (start_temp_num <= test_dataset_modified['subject_idx'][x][1] ) & (test_dataset_modified['subject_idx'][x][1] <= end_temp_num) :
                        modified_sentence += temp
                        temp = "" 
                    elif (start_temp_num <= test_dataset_modified['object_idx'][x][0] ) & (test_dataset_modified['object_idx'][x][0] <= end_temp_num) :
                        modified_sentence += temp
                        temp = ""  
                    elif (start_temp_num <= test_dataset_modified['object_idx'][x][1] ) & (test_dataset_modified['object_idx'][x][1] <= end_temp_num) :
                        modified_sentence += temp
                        temp = "" 
                    else : 
                        temp = ""
                elif 괄호cnt < 0 :
                    broken_cnt.append(x)
                    
            # 괄호에 해당되지 않는 문자열이면
            else :
                if 괄호cnt != 0 :
                    temp += test_dataset_modified['sentence'][x][i]
                else :
                    modified_sentence += test_dataset_modified['sentence'][x][i] 
    
    if len(modified_sentence) >= 1 : 
        test_dataset_modified['sentence'][x] = modified_sentence
    else : 
        pass

broken_cnt = set(broken_cnt)
 
```

```python
data= dataset
        
sub_entity, sub_type= [], []
obj_entity, obj_type= [], []
sub_idx, obj_idx= [], []
sentence= []

"""preprocess"""
for i, [x, y, z] in enumerate(zip(data['subject_entity'], data['object_entity'], data['sentence'])):
    sub_typ= x[1:-1].split(':')[-1].split('\'')[-2]
    obj_typ= y[1:-1].split(':')[-1].split('\'')[-2]
            
    for idx_i in range(len(x)):
        if x[idx_i: idx_i+ 9]== 'start_idx':
            sub_start= int(x[idx_i+12:].split(',')[0].strip())
        if x[idx_i: idx_i+7]== 'end_idx':
            sub_end= int(x[idx_i+10:].split(',')[0].strip())
                
        if y[idx_i: idx_i+ 9]== 'start_idx':
            obj_start= int(y[idx_i+12:].split(',')[0].strip())
        if y[idx_i: idx_i+7]== 'end_idx':
            obj_end= int(y[idx_i+10:].split(',')[0].strip())
            
    sub_i= [sub_start, sub_end]
    obj_i= [obj_start, obj_end]

    sub_entity.append(z[sub_i[0]: sub_i[1]+1])
    obj_entity.append(z[obj_i[0]: obj_i[1]+1])
    sub_type.append(sub_typ); sub_idx.append(sub_i)
    obj_type.append(obj_typ); obj_idx.append(obj_i)
    
    if sub_i[0] < obj_i[0]:
        z= z[:sub_i[0]] + '@*'+sub_typ+'*'+ z[sub_i[0]: sub_i[1]+1] + '@' + z[sub_i[1]+1:]
        z= z[:obj_i[0]+7] + '#^'+ obj_typ +'^'+ z[obj_i[0]+7: obj_i[1]+8]+ '#'+ z[obj_i[1]+8:]
    else:
        z= z[:obj_i[0]] + '#^'+ obj_typ +'^'+ z[obj_i[0]: obj_i[1]+1]+ '#' + z[obj_i[1]+1:]
        z= z[:sub_i[0]+7] + '@*'+sub_typ+'*' + z[sub_i[0]+7: sub_i[1]+8] + '@' + z[sub_i[1]+8:]
        
    sentence.append(z)

test_df= pd.DataFrame({'id': data['id'], 'sentence' : sentence, 'subject_entity': sub_entity, 'object_entity': obj_entity,
                        'subject_type': sub_type, 'object_type': obj_type, 'label': data['label'],
                        'subject_idx': sub_idx, 'object_idx': obj_idx})
```

```python
from transformers import AutoTokenizer, AutoModel, Trainer, TrainingArguments, EarlyStoppingCallback



def tokenized_dataset(data):
    tokenizer= AutoTokenizer.from_pretrained('klue/roberta-large')
    """add token list"""
    # Dataset에는 PER(사람), ORG(조직), DAT(시간), LOC(장소), POH(기타 표현), NOH(기타 수량 표현) 총 6가지 Type이 존재
    tokens= ['PER', 'LOC', 'POH', 'DAT', 'NOH', 'ORG']
    tokenizer.add_tokens(tokens)     

    concat_entity = []
    for sub_ent, obj_ent, sub_typ, obj_typ in zip(data['subject_entity'], data['object_entity'], data['subject_type'], data['object_type']):
        temp =  '@*'+ sub_typ + '*' + sub_ent + '@와 #^' + obj_typ + '^' + obj_ent + '#의 관계'
        #temp =  e01 + '와' + e02 + '의 관계'
        concat_entity.append(temp)
        
    print(concat_entity[:10])
    
    tokenized_sentence= tokenizer(
        concat_entity,
        list(data['sentence']), # list나 string type으로 보내줘야 함 !
        return_tensors= "pt", # pytorch type
        padding= True, # 문장의 길이가 짧다면 padding
        truncation= True, # 문장 자르기
        max_length= 256, # 토큰 최대 길이...
        add_special_tokens= True, # special token 추가
        return_token_type_ids= False # roberta의 경우.. token_type_ids가 안들어감 ! 
    )    

    return tokenized_sentence, len(tokens)
```
